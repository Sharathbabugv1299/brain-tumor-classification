```python
import os
from PIL import Image
from matplotlib import pyplot as plt
import shutil
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from skimage import io

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.utils.class_weight import compute_class_weight
import tensorflow as tf
```

```python
# Define data paths
data_dir = 'Dataset'
healthy_dir = os.path.join(data_dir, 'Healthy')
tumor_dir = os.path.join(data_dir, 'Brain Tumor')

# Function to load dataset and print number of images
def load_dataset_info():
    healthy_images = os.listdir(healthy_dir)
    tumor_images = os.listdir(tumor_dir)

    print(f'Number of healthy images: {len(healthy_images)}')
    print(f'Number of brain tumor images: {len(tumor_images)}')
    
    return healthy_images, tumor_images

# Load dataset information
healthy_images, tumor_images = load_dataset_info()
```

```python
import matplotlib.pyplot as plt
from PIL import Image

# Function to plot sample images from each class
def plot_sample_images(images, category, folder_dir):
    plt.figure(figsize=(15, 10))
    for i in range(8):  # Plotting 9 sample images
        plt.subplot(4, 4, i + 1)
        img_path = os.path.join(folder_dir, images[i])
        img = Image.open(img_path)
        plt.imshow(img, cmap='gray')
        plt.title(f'{category}')
        plt.axis('off')
    plt.show()

# Visualize samples from Healthy class
plot_sample_images(healthy_images, "Healthy", healthy_dir)

# Visualize samples from Brain Tumor class
plot_sample_images(tumor_images, "Brain Tumor", tumor_dir)
```

```python
# Function to analyze image sizes
def analyze_image_sizes(images, folder_dir):
    sizes = []
    for img_name in images:
        img_path = os.path.join(folder_dir, img_name)
        img = Image.open(img_path)
        sizes.append(img.size)  # Store image sizes as (width, height) tuples
    return sizes

# Analyze image sizes for both classes
healthy_sizes = analyze_image_sizes(healthy_images, healthy_dir)
tumor_sizes = analyze_image_sizes(tumor_images, tumor_dir)

# Function to plot the distribution of image sizes
import seaborn as sns

def plot_image_size_distribution(healthy_sizes, tumor_sizes):
    all_sizes = healthy_sizes + tumor_sizes
    width, height = zip(*all_sizes)  # Unpack sizes into width and height

    plt.figure(figsize=(10, 6))
    sns.histplot(width, kde=True, label='Width', color='blue')
    sns.histplot(height, kde=True, label='Height', color='red')
    plt.title('Distribution of Image Sizes')
    plt.xlabel('Pixels')
    plt.legend()
    plt.show()

plot_image_size_distribution(healthy_sizes, tumor_sizes)
```

```python
def check_image_sizes(folder):
    sizes = []
    for filename in os.listdir(folder):
        img_path = os.path.join(folder, filename)
        if os.path.isfile(img_path):
            img = Image.open(img_path)
            sizes.append(img.size)
    return sizes

tumor_sizes = check_image_sizes(tumor_dir)
healthy_sizes = check_image_sizes(healthy_dir)

print("Tumor image sizes:", set(tumor_sizes))
print("Healthy brain image sizes:", set(healthy_sizes))
```

```python
def plot_intensity_distribution(class_dir, num_images=200):
    # class_dir = os.path.join(data_dir, class_name)
    images = [io.imread(os.path.join(class_dir, img)) for img in os.listdir(class_dir)[:num_images]]
    
    plt.figure(figsize=(10, 5))
    for i, img in enumerate(images):
        plt.hist(img.ravel(), bins=256, alpha=0.5, label=f'Image {i+1}')
    
    plt.title(f'Pixel Intensity Distribution - {class_dir} class')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')
    plt.show()

# Plot intensity distribution for each class
for class_name_dir in [healthy_dir, tumor_dir]:
    plot_intensity_distribution(class_name_dir)
```

```python
def calculate_intensity_stats(images, folder_dir):
    means, std_devs = [], []
    
    for img_name in images:
        img_path = os.path.join(folder_dir, img_name)
        img = Image.open(img_path)
        img_array = np.array(img)
        
        # Calculate mean and standard deviation of pixel intensities
        means.append(np.mean(img_array))
        std_devs.append(np.std(img_array))
    
    # Return the overall mean and standard deviation
    return np.mean(means), np.mean(std_devs)

# Calculate stats for Healthy and Brain Tumor images
healthy_mean, healthy_std = calculate_intensity_stats(healthy_images, healthy_dir)
tumor_mean, tumor_std = calculate_intensity_stats(tumor_images, tumor_dir)

print(f'Healthy images - Mean Intensity: {healthy_mean:.2f}, Standard Deviation: {healthy_std:.2f}')
print(f'Brain Tumor images - Mean Intensity: {tumor_mean:.2f}, Standard Deviation: {tumor_std:.2f}')
```

```python
import os
from PIL import Image

# Resize images and handle RGBA conversion
def resize_images(source_folder, target_folder, target_size=(224, 224)):
    if not os.path.exists(target_folder):
        os.makedirs(target_folder)
    
    for img_name in os.listdir(source_folder):
        img_path = os.path.join(source_folder, img_name)
        
        try:
            # Open image
            img = Image.open(img_path)

            # Convert 'RGBA' or 'L' images to 'RGB' (JPEG does not support 'RGBA')
            if img.mode == 'RGBA' or img.mode == 'L':
                img = img.convert('RGB')

            # Resize image
            img_resized = img.resize(target_size)
            
            # Save the image in JPEG format, even if the original is TIFF or another format
            save_path = os.path.join(target_folder, os.path.splitext(img_name)[0] + '.jpg')
            img_resized.save(save_path, 'JPEG')
        
        except Exception as e:
            print(f"Error processing image {img_name}: {e}")

# Resize images in both Healthy and Brain Tumor folders
resize_images(healthy_dir, 'Preprocessed_Draft/Healthy', target_size=(224, 224))
resize_images(tumor_dir, 'Preprocessed_Draft/Brain Tumor', target_size=(224, 224))
```

```python
import os
import shutil
import random

def create_dir_structure(base_dir):
    """Create directory structure for train, val, and test sets."""
    if not os.path.exists(base_dir):
        os.makedirs(base_dir)

    for split in ['train', 'val', 'test']:
        for category in ['Brain Tumor', 'Healthy']:
            split_path = os.path.join(base_dir, split, category)
            os.makedirs(split_path, exist_ok=True)

def split_data(source_dir, dest_dir, train_size=0.7, val_size=0.15, test_size=0.15):
    """
    Randomly split the dataset into train, validation, and test sets.
    
    Parameters:
    - source_dir: Path to the original dataset.
    - dest_dir: Path where the split dataset will be saved.
    - train_size: Proportion of data to be used for training (default 70%).
    - val_size: Proportion of data to be used for validation (default 15%).
    - test_size: Proportion of data to be used for testing (default 15%).
    """

    assert train_size + val_size + test_size == 1, "Train, validation, and test sizes must sum to 1."

    create_dir_structure(dest_dir)

    # Iterate through categories (Brain Tumor, Healthy)
    for category in ['Brain Tumor', 'Healthy']:
        category_path = os.path.join(source_dir, category)
        images = os.listdir(category_path)
        
        # Shuffle the images to ensure random splitting
        random.shuffle(images)

        # Split indices for train, validation, and test sets
        train_split = int(train_size * len(images))
        val_split = int((train_size + val_size) * len(images))

        # Split into train, validation, and test sets
        train_images = images[:train_split]
        val_images = images[train_split:val_split]
        test_images = images[val_split:]

        # Copy images to the corresponding directories
        for image in train_images:
            src_path = os.path.join(category_path, image)
            dest_path = os.path.join(dest_dir, 'train', category, image)
            shutil.copy(src_path, dest_path)

        for image in val_images:
            src_path = os.path.join(category_path, image)
            dest_path = os.path.join(dest_dir, 'val', category, image)
            shutil.copy(src_path, dest_path)

        for image in test_images:
            src_path = os.path.join(category_path, image)
            dest_path = os.path.join(dest_dir, 'test', category, image)
            shutil.copy(src_path, dest_path)

    print("Data split complete!")

# Example usage
source_dataset_dir = "Preprocessed_Draft"  # Path to your dataset folder
destination_dataset_dir = "Dataset_Split_Draft"  # Where the split data will be stored

split_data(source_dataset_dir, destination_dataset_dir, train_size=0.75, val_size=0.15, test_size=0.10)
```

### Base model

```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Directories for your data
train_dir = 'Dataset_Split_Draft/Train'
val_dir = 'Dataset_Split_Draft/Val'
test_dir = 'Dataset_Split_Draft/Test'
```

#### VGG16 Base Model

```python

# Load pre-trained VGG16 model without top layers
base_model = tf.keras.applications.VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Freeze all layers of the base model
for layer in base_model.layers:
    layer.trainable = False

# Add custom layers on top of VGG16
x = base_model.output
x = GlobalAveragePooling2D()(x)
predictions = Dense(1, activation='sigmoid')(x)  # Binary classification

# Create the model
model = Model(inputs=base_model.input, outputs=predictions)

# Compile the model
optimizer = Adam(learning_rate=1e-4)
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

# Data generators without augmentation
train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

# Load the data
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    shuffle=False
)

# Train the model
history = model.fit(
    train_generator,
    epochs=15,
    validation_data=val_generator,
    steps_per_epoch=len(train_generator),
    validation_steps=len(val_generator)
)

# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(test_generator, steps=len(test_generator))
print(f'Test loss: {test_loss}')
print(f'Test accuracy: {test_accuracy}')

# Predictions on the test set
y_pred_probs = model.predict(test_generator)
y_pred = np.round(y_pred_probs).astype(int)
y_true = test_generator.classes

# Print classification report
print(classification_report(y_true, y_pred, target_names=['Healthy', 'Brain Tumor']))

# Confusion matrix
conf_matrix = confusion_matrix(y_true, y_pred)

# Plot confusion matrix
plt.figure(figsize=(8,6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Healthy', 'Brain Tumor'], yticklabels=['Healthy', 'Brain Tumor'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

### Enhanced model

#### Resnet50

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.applications.resnet50 import preprocess_input

# Load pre-trained ResNet50 model + higher level layers
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Freeze the bottom layers of ResNet50 model and unfreeze the top layers for fine-tuning
for layer in base_model.layers[:143]:
    layer.trainable = False
for layer in base_model.layers[143:]:
    layer.trainable = True

# Adding custom layers on top of ResNet50 base model
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(512, activation='relu')(x)
x = BatchNormalization()(x)  # Add Batch Normalization
x = Dropout(0.5)(x)
predictions = Dense(1, activation='sigmoid')(x)

# Create the final model
model = Model(inputs=base_model.input, outputs=predictions)

# Compile the model
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001), loss='binary_crossentropy', metrics=['accuracy'])

# Data augmentation for training
train_datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.3,
    height_shift_range=0.3,
    shear_range=0.3,
    zoom_range=0.3,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='nearest',
    preprocessing_function=preprocess_input  # Preprocessing for ResNet50
)

# Data generator for validation and test (without augmentation, but with preprocessing)
val_test_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input
)

# Load training data
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    shuffle=True
)

# Load validation data
val_generator = val_test_datagen.flow_from_directory(
    val_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    shuffle=False
)

# Load test data
test_generator = val_test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    shuffle=False
)

# Implement early stopping and learning rate reduction
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=0.00001)

# Train the model with early stopping and learning rate reduction
history = model.fit(
    train_generator,
    epochs=30,
    verbose=1,
    callbacks=[early_stopping, reduce_lr],
    validation_data=val_generator  # Use separate validation data generator
)

# Model evaluation
test_loss, test_accuracy = model.evaluate(test_generator, verbose=1)
print(f'Test loss: {test_loss}')
print(f'Test accuracy: {test_accuracy}')

# Predict on the test set
test_generator.reset()  # Reset the generator to ensure correct predictions
y_pred = model.predict(test_generator, steps=len(test_generator), verbose=1)
y_pred = np.round(y_pred).astype(int)  # Convert predictions to binary

# True labels
y_true = test_generator.classes

# Compute confusion matrix
cm = confusion_matrix(y_true, y_pred)
print("Confusion Matrix:\n", cm)

# Compute classification report
report = classification_report(y_true, y_pred, target_names=['Healthy', 'Brain Tumor'])
print("Classification Report:\n", report)

# Plot training & validation accuracy and loss
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.show()

# Plot Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Healthy', 'Brain Tumor'], yticklabels=['Healthy', 'Brain Tumor'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Model

# Function to extract features using the model's intermediate layer
def extract_features(model, generator):
    intermediate_layer_model = Model(inputs=model.input, outputs=model.layers[-2].output)  # Extract features before the last layer
    features = intermediate_layer_model.predict(generator)
    return features

class_labels = ['Brain Tumor', 'Healthy']

# Create a data generator for the validation set
test_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    shuffle=False
)

# Extract features from the validation set
features = extract_features(model, test_generator)
labels = test_generator.classes  # Get the true labels of the validation data

# Apply t-SNE to reduce dimensions
tsne = TSNE(n_components=2, random_state=42)  # 2D t-SNE
features_tsne = tsne.fit_transform(features)

# Plotting the t-SNE result
plt.figure(figsize=(10, 7))

# Plot each class with a different color
for label in np.unique(labels):
    indices = np.where(labels == label)
    plt.scatter(features_tsne[indices, 0], features_tsne[indices, 1], label=class_labels[int(label)], alpha=0.5)

plt.legend()
plt.title('t-SNE Visualization of Image Features')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.show()
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.models import Model
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, BatchNormalization

# Step 1: Define the model architecture (same as the model you used for training)
base_model = ResNet50(weights=None, include_top=False, input_shape=(224, 224, 3))

# Freeze the bottom layers of ResNet50 model and unfreeze the top layers for fine-tuning
for layer in base_model.layers[:143]:
    layer.trainable = False
for layer in base_model.layers[143:]:
    layer.trainable = True

# Adding custom layers on top of ResNet50 base model
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(512, activation='relu')(x)
x = BatchNormalization()(x)
x = Dropout(0.5)(x)
predictions = Dense(1, activation='sigmoid')(x)

# Step 2: Create the final model and load the saved weights
model = Model(inputs=base_model.input, outputs=predictions)

# Path to the saved weights file
saved_weights_path = 'best_model_weights.h5'  # Update with your actual weights path
model.load_weights(saved_weights_path)

# Step 3: Define the image size and class labels
IMG_SIZE = (224, 224)
class_labels = ['Brain Tumor', 'Healthy']

# Function to load and preprocess the image
def load_and_preprocess_image(img_path):
    img = load_img(img_path, target_size=IMG_SIZE)
    img_array = img_to_array(img)
    img_preprocessed = preprocess_input(np.expand_dims(img_array, axis=0))
    return img_preprocessed

# Function to predict the class of an image
def predict_image(img_path, model):
    img_preprocessed = load_and_preprocess_image(img_path)
    prediction = model.predict(img_preprocessed)
    predicted_label = class_labels[int(prediction[0] > 0.5)]  # Binary classification
    return predicted_label

# Folder path containing images to predict (replace with your folder)
image_folder = 'junk/unseen_data'  # Replace with your test folder

# Predict and display images
def predict_and_display_images(folder_path, model):
    images = os.listdir(folder_path)
    
    plt.figure(figsize=(12, 12))
    
    for i, img_name in enumerate(images):
        img_path = os.path.join(folder_path, img_name)
        
        # Make prediction
        predicted_label = predict_image(img_path, model)
        
        # Actual label logic based on filename or folder (can customize this)
        actual_label = "Healthy" if 'Not Cancer' in img_name else "Brain Tumor"

        # Load the image for display
        img = load_img(img_path, target_size=IMG_SIZE)
        
        # Display the image with actual and predicted labels
        plt.subplot(4, 4, i + 1)  # Display images in a grid (4x4 here)
        plt.imshow(img)
        plt.title(f"Actual: {actual_label}\nPredicted: {predicted_label}")
        plt.axis('off')  # Turn off the axis

        # Limit to displaying 16 images, adjust as needed
        if i >= 15:
            break

    plt.tight_layout()
    plt.show()

# Call the function to predict and display images
predict_and_display_images(image_folder, model)
```
